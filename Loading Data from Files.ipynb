## JSON

```python
import json

data = {"name": "Alfredo", "data": [], "valid": True}

# convert python object (dictionary above) to the json string
json.dumps(data)
```

```python
json_output = '{"name": "Alfredo", "data": 1, "valid": true}'

# convert json to python object (dictionary in this case)
loaded_json = json.loads(json_output)
print(type(loaded_json))
print(loaded_json)

loaded_json['valid']
```
